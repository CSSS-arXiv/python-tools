```python
def ReturnTime(s):
    """ s is a string-formatted name of a paper
    ReturnTime returns a 6-digit time stamp YYYYMM """
    if s.count('.') == 1:
        if s[0] == '9':
            return int('19' + s[0:4])
        else:
            return int('20' + s[0:4])
    elif s.count('/') == 1:
        s1 = s.partition('/')[-1]
        if s1[0] == '9':
            return int('19' + s1[:4])
        else:
            return int('20' + s1[:4])
```

```python
network_papers = np.array([p[0] for p in data])
network_papers[20:]
```

```python
# now we create the networks of authors
# - keep track of time when papers were added

# {timestamp: [list of papers added at timestamp]}
tpapers = defaultdict(list) 
for paper in network_papers:
    tstamp = ReturnTime(paper)
#    if tstamp < 195000: print paper, tstamp
    tpapers[tstamp].append(paper)
```

```python
# - keep track of the time when NEW authors were added

# {author: [time stamps]} -> {author: minimum time stamps}
auth_tstamps = defaultdict(list) 
for paper in network_papers:
    for author in padict[paper]:
        # find the list of all papers by the author also in the set of network_papers
        aps = set(apdict[author]) & set (network_papers)
        if author not in auth_tstamps:
            # find the minimum time stamp, looping through all papers
            auth_tstamps[author] = min([ReturnTime(p) for p in aps])

# {time stamp : [authors]}
tauthors = defaultdict(list) 
for author, t in auth_tstamps.iteritems():
    tauthors[t].append(author)
tauthors = dict(tauthors)            
```

# Create networks of authors on the networks topic

This is where the magic happens

```python
# What do we need here? 
# - we have the times when each author was added
# - we have the times when each paper was added
# - we have all the authors per paper
```

```python
tstamps = sorted(tpapers.keys()); # all relevant time stamps
tnodes = defaultdict(list) # {timestamp: [cumulative list of nodes]}
tedges = defaultdict(list) # {timestamp: [cumulative list of edges]}
for ti in tstamps:
    all_auths = []
    all_edges = []
    for paper in tpapers[ti]:
        if paper in padict:
            # list of authors for the paper
            auths = padict[paper] 
            # list of edges, including multi-edges in case we want to count number of collaborations later
            edges = [sorted(pair) for pair in itertools.combinations(auths, 2)]
            all_auths.extend(auths)
            all_edges.extend(edges)
    tnodes[ti] = all_auths
    tedges[ti] = all_edges
```

# Make graphs of following metrics:

1. Number of edges/number of nodes
2. Proportion of edges/nodes in largest component
3. Proportion of edges/nodes in 2nd largest component
4. Proportion of edges/nodes in largest bi-component (avoids dumbell-shaped communities)
5. Diameter of gcc, bicomponent - average shortest path length

Note: triadic closure is another measurable quantity, can look into later if other group is interested

Note: the inconsistency in counting number of participating authors I was finding before was that tnodes was double-counting certain authors across time steps, needed to use the set of all participating authors for a consistent number

```python
test = nx.Graph()
nnodes = [] # number of nodes in full graph
nedges = [] # number of edges in full graph
comp1_sizes = [] # [number of nodes, number of edges] in largest component
comp2_sizes = [] # [number of nodes, number of edges] in second-largest component
bicomp_sizes = [] # [number of nodes, number of edges] in largest bicomponent
diameters = [] # [diameter of gcc, diameter of largest bicomp]
for ti in tstamps:
    # retrieve lists of nodes and edges
    nodes = tnodes[ti]
    edges = tedges[ti]
    # construct graph
    for auth in nodes:
        test.add_node(auth)
    for edge in edges:
        test.add_edge(*edge)
    # counting all nodes and edges
    nnodes.append(len(test.nodes()))
    nedges.append(len(test.edges()))
    # look at first two largest connected components
    sgs = nx.connected_component_subgraphs(test)
    comp1_sizes.append([len(sgs[0]),len(sgs[0].edges())])
    if len(sgs) > 1:
        comp2_sizes.append([len(sgs[1]),len(sgs[1].edges())])
    else:
        comp2_sizes.append([0,0])
    # look at largest bicomponent
    bcs = nx.biconnected_component_subgraphs(test).next()
    bicomp_sizes.append([len(bcs), len(bcs.edges())])
    # measure diameters
    d1 = nx.average_shortest_path_length(sgs[0])
    d2 = nx.average_shortest_path_length(bcs)
    diameters.append([d1, d2])
diameters = np.array(diameters).transpose()
comp1_sizes = np.array(comp1_sizes).transpose()
comp2_sizes = np.array(comp2_sizes).transpose()
bicomp_sizes = np.array(bicomp_sizes).transpose()
```

```python
# Plot how number of edges scales with number of nodes over time in full network
plt.plot(nnodes, nedges, linewidth = 2)
x = y = logspace(0,4,100)
plt.plot(x,.5*y, linewidth = 2)
loglog()
xlim(1,10**4)
ylim(1,10**4)
legend(['Edges vs. nodes', 'Linear growth'], loc = 'upper left', fontsize = 14)
plt.title('Topic: Networks', fontsize = 16)
xlabel('Number of nodes', fontsize = 14)
ylabel('Number of (unique) edges', fontsize = 14)
#savefig('arxiv_figures/network_edge_scaling.pdf')
```

```python
# Plot proportion of edges in largest connected component
plt.plot(nnodes, 1.*comp1_sizes[0]/nnodes, linewidth = 2)
plt.plot(nnodes, 1.*comp2_sizes[0]/nnodes, linewidth = 2)
ylabel('Fraction of nodes in \n largest connected component', fontsize = 14)
xlabel('Number of nodes', fontsize = 14)
plt.title('Topic: Networks' , fontsize = 16);
legend(['Largest component', '2nd-largest component'], loc = 'upper right', fontsize=14)
#savefig('arxiv_figures/network_gcc_nodes.pdf')
```

```python
# Plot proportion of edges in largest connected component
plt.plot(nnodes, 1.*comp1_sizes[1]/nedges, linewidth = 2)
plt.plot(nnodes, 1.*comp2_sizes[1]/nedges, linewidth = 2)
ylabel('Fraction of edges in \n largest connected component', fontsize = 14)
xlabel('Number of nodes', fontsize = 14)
plt.title('Topic: Networks' , fontsize = 16);
legend(['Largest component', '2nd-largest component'], loc = 'upper right', fontsize=14)
#savefig('arxiv_figures/network_gcc_edges.pdf')
```

```python
# Plot scaling of edges/nodes inside the largest connected component
plt.plot(1.*comp1_sizes[0], 1.*comp1_sizes[1], linewidth = 2)
xlabel('Number of nodes in LCC', fontsize = 14)
ylabel('Number of (unique) edges in LCC', fontsize = 14)
plt.title('Topic: Networks' , fontsize = 16);
loglog()
x = y = logspace(0,4,100)
plt.plot(x,.5*y, linewidth = 2)
xlim(1,10**4)
ylim(1,10**4)
legend(['Largest CC Edges vs. nodes', 'Linear growth'], loc = 'upper left');
```

```python
# Plot fraction of nodes in largest bicomponent
plt.plot(1.*bicomp_sizes[0]/nnodes, linewidth = 2)
plt.plot(1.*bicomp_sizes[1]/nedges, linewidth = 2)
legend(['Fraction of nodes in \n largest connected biocomponent', 'Fraction of edges in \n largest connected bicomponent'], fontsize = 14, loc = 'upper right')
xlabel('Number of Nodes', fontsize = 14)
ylim(0,1.1)
plt.title('Topic: Networks' , fontsize = 16);
#savefig('arxiv_figures/network_bicomp.pdf')
```

```python
# diameters, mean geodesic path
plt.plot(nnodes, diameters[0], linewidth = 2)
plt.plot(nnodes, diameters[1], linewidth = 2)
legend(['Largest connected component', 'Largest connected bicomponent'], loc='upper right')
ylim(1,12)
ylabel('Mean Geodesic Path Length', fontsize = 14)
xlabel('Number of nodes', fontsize = 14)
plt.title('Topic: Networks', fontsize = 16)
#savefig('arxiv_figures/network_diameters.pdf')
```
